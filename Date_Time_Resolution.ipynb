```python
import pandas as pd
from datetime import datetime, timedelta
import math
df = pd.read_csv(r'E:\graph-fraud-detection-main\ieee-data\train_transaction.csv')

start_date = pd.Timestamp('2017-11-01')
class TimeConverter:
    """
    turn TransactionDT seconds into real dates.
    """
    def __init__(self, start_date="2017-11-01"):
        # starting point 
        self.start = pd.Timestamp(start_date)
        print(f"Time machine set to {self.start}!")

    def seconds_to_date(self, seconds):
        """
        Converts seconds to a date using timedelta.
         s = t0 + Δt where Δt is seconds
        """
        
        if isinstance(seconds, (int, float)):
            # Δt = seconds, so new time = start + Δt
            delta = timedelta(seconds=math.floor(seconds))
            new_date = self.start + delta
            return new_date
        elif isinstance(seconds, pd.Series):
            return pd.to_datetime(self.start + pd.to_timedelta(seconds, unit='s'))
        else:
            print("Error: Input must be a number or pandas Series!")
            return None

if __name__ == "__main__":
    
    converter = TimeConverter(start_date="2017-11-01")
    
    # Convert TransactionDT to dates (solving for t in t = t0 + Δt)
    df['RealDate'] = converter.seconds_to_date(df['TransactionDT'])
    

    print("First few rows with converted dates:")
    print(df[['TransactionID', 'TransactionDT', 'RealDate']].head())
```

```python
df['TransactionDateTime'] = pd.to_datetime(start_date + pd.to_timedelta(df['TransactionDT'], unit='s'))

df['TransactionDate'] = df['TransactionDateTime'].dt.date

print(df[['TransactionID', 'TransactionDT', 'TransactionDateTime', 'TransactionDate']].head())

```

```python
df.to_csv('train_transaction_with_dates.csv', index=False)
```
